__Задание 1.__   
Доказать $(\pmb{AB})^{\top} = \pmb{B}^{\top}\pmb{A}^{\top}$.

__Решение:__
$$(\pmb{AB})^{\top} = \left(\begin{pmatrix}
a_{11} & \dots & a_{1m} \\
\vdots & \ddots & \vdots \\
a_{n1}& \dots & a_{nm}
\end{pmatrix} \times \begin{pmatrix}
b_{11} & \dots & b_{1m} \\
\vdots & \ddots & \vdots \\
b_{n1}& \dots & b_{nm} \end{pmatrix} \right)^\top = 
\begin{pmatrix} 
a_{11}b_{11}+\dots +a_{1m}b_{n1} & \dots & a_{11}b_{1m}+\dots +a_{1m}b_{nm} \\
\vdots & \ddots & \vdots \\
a_{n1}b_{11}+\dots +a_{nm}b_{n1} & \dots & a_{n1}b_{1m}+\dots +a_{nm}b_{nm}
\end{pmatrix}^\top = \begin{pmatrix} 
a_{11}b_{11}+\dots +a_{1m}b_{n1} & \dots & a_{n1}b_{11}+\dots +a_{nm}b_{n1} \\
\vdots & \ddots & \vdots \\
a_{11}b_{1m}+\dots +a_{1m}b_{nm} & \dots & a_{n1}b_{1m}+\dots +a_{nm}b_{nm}
\end{pmatrix} (1)$$

$$ \pmb{B}^{\top}\pmb{A}^{\top} = \begin{pmatrix}
b_{11} & \dots & b_{1m} \\
\vdots & \ddots & \vdots \\
b_{n1}& \dots & b_{nm} \end{pmatrix}^\top \times \begin{pmatrix}
a_{11} & \dots & a_{1m} \\
\vdots & \ddots & \vdots \\
a_{n1}& \dots & a_{nm}
\end{pmatrix}^\top  = \begin{pmatrix}
b_{11} & \dots & b_{n1} \\
\vdots & \ddots & \vdots \\
b_{1m}& \dots & b_{nm} \end{pmatrix} \times \begin{pmatrix}
a_{11} & \dots & a_{n1} \\
\vdots & \ddots & \vdots \\
a_{1m}& \dots & a_{nm}
\end{pmatrix} = \begin{pmatrix} 
a_{11}b_{11}+\dots +a_{1m}b_{n1} & \dots & a_{n1}b_{11}+\dots +a_{nm}b_{n1} \\
\vdots & \ddots & \vdots \\
a_{11}b_{1m}+\dots +a_{1m}b_{nm} & \dots & a_{n1}b_{1m}+\dots +a_{nm}b_{nm}
\end{pmatrix} (2) $$

Поскольку (1) = (2), $(\pmb{AB})^{\top} = \pmb{B}^{\top}\pmb{A}^{\top}$ . 

__Что и требовалось доказать__ .

---

__Определение.__ Матрица $\pmb{M}$ называется ортогональной, если выполняется следующее тождество:
$$
\pmb{M M}^\top = \pmb{M}^\top \pmb{M} = \pmb{I}
$$  
где $\pmb{I}$ - единичная матрица. Другими словами, для ортогональной матрицы верно:  
$$
\pmb{M}^{-1} = \pmb{M}^\top
$$

__Задание 2.__   
$\pmb{A}$ и $\pmb{B}$ ортогональны. Покажите равносильность всех следующих утверждений между собой:

1. $\pmb{AB} = \pmb{BA}$
2. $\pmb{A}^\top \pmb{B} = \pmb{BA}^\top$
3. $\pmb{AB}^\top = \pmb{B}^\top \pmb{A}$
4. $\pmb{A}^\top \pmb{B}^\top = \pmb{B}^\top \pmb{A}^\top$

Подсказка: утверждения равносильные, если от одного к другому можно перейти при помощи набора некоторых операций. Например, здесь можно перейти от первого ко второму, от второго к третьему, от третьего к четвертому и от четвертого к первому.

__Решение__:

$1 \Rightarrow 2$ : $AB=BA$; $B=A^TBA$;  $BA^T=A^TB$

$2 \Rightarrow 3$ : $A^TB=BA^T$; $(B^TA)^T=(AB^T)^T$; $((B^TA)^T)^T=((AB^T)^T)^T \Rightarrow AB^T=B^TA$

$3 \Rightarrow 4$ : $AB^T=B^TA$; $B^T=A^TB^TA$; $B^TA^T=A^TB^T$

$4 \Rightarrow 1$ : $A^TB^T=B^TA^T$; $(A^TB^T)^T=(B^TA^T)^T$; $BA=AB$

Все утверждения эквивалентны, __что и требовалось доказать__.

__Задание 1.__ 

Дана следующая квадратная матрица:
$$\begin{pmatrix}
2 & -1 &  &  & 0 \\
-1 & 2 & -1 &  & \\
& \ddots & \ddots & \ddots & \\
& & -1 & 2 & -1\\
0& & & -1 & 2
\end{pmatrix}$$  
На главной диагонали значения равны `2`, на диагоналях на 1 выше и на 1 ниже значения равны `-1`. __Все__ остальные значения равны `0`.    
Необходимо создать такую матрицу с помощью `numpy` и без использования циклов.

````python
def tridiagonal_matrix(n):
    return 2 * np.eye(n) - np.eye(n, k=1) - np.eye(n, k=-1)

tridiagonal_matrix(4)
````

__Задание 2.__

Посчитать [матрицу Гильберта](https://ru.wikipedia.org/wiki/%D0%9C%D0%B0%D1%82%D1%80%D0%B8%D1%86%D0%B0_%D0%93%D0%B8%D0%BB%D1%8C%D0%B1%D0%B5%D1%80%D1%82%D0%B0) размера $n \times n$ без использования циклов.  
Подсказка: можно использовать функцию `np.fromfunction`.

````python
# float
def hilbert(n):
    matrix = np.fromfunction(lambda i, j: 1 / (i + j + 1), (n, n))
    return matrix

hilbert(4)
````

````python
# Fraction
from fractions import Fraction

def hilbert(n):
    inv = np.vectorize(lambda x: Fraction(1, x))
    return inv(np.add.outer(np.arange(0, n),
                            np.arange(1, n+1)))
hilbert(4)
````

Подсказки:
* Напишите функцию, которая вычисляет ядро Гаусса заданного размера `shape`:

```python
def gaussian(shape, mean_x, mean_y, sigma_x, sigma_y):
    '''
    @param: shape: 2d size of output kernel
    '''
    gauss_kernel = np.empty(shape)
    # ...

# gaussian(x, y, ...).shape == (n, m)
```

* Сгенерируйте ядро Гаусса с помощью указанной функции со средними значениями $mean_x=0$ и $mean_y=0$, затем нормируйте полученную матрицу, чтобы значения получились в диапазоне $[0, 1]$.
* Умножьте полученную матрицу на исходное изображение.
* Варьируйте значения $mean_x$ и $mean_y$ для получения наилучшего результата. Как сигмы можно взять следующие значения:  
$\sigma_x$ = `img.shape[0] // 2`, $\sigma_y$ = `img.shape[1] // 2`.

Ожидаемый результат должен выглядеть следующим образом:

![Лемур виньетка](data/lemur_vignette.png)

__Решение__:

````python
from skimage.io import imread, imshow_collection
from pylab import rcParams
%matplotlib inline


def gaussian(shape, mean_x, mean_y, sigma_x, sigma_y):
    a = np.fromfunction(lambda y, x, k: -((x-mean_x) ** 2 / (2*sigma_x**2) + \
                                          (y-mean_y) ** 2 / (2*sigma_y**2)), shape)
    return np.exp(a)


img = imread('data/lemur.jpg')

a = gaussian(img.shape, 1400, 400, img.shape[0] // 2, img.shape[1] // 2)

copy = np.multiply(img, a).astype(np.uint8)
    
rcParams['figure.figsize'] = (15, 15)    
imshow_collection([img, copy])
````
